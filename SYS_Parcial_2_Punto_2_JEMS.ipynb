```python
import numpy as np
import matplotlib.pyplot as plt
#Frecuencia portadora y muestreo

Fc = 100 #frecuencia de portadora

Fm = 10 #frecuencia del mensaje
Am = 1 #amplitud del mensaje


ti = 0
tf = 5/Fm #cantidad de periodos a graficar de la

Fs = 5*Fc #frecuencia muestreo

t = np.arange(ti,tf,1/Fs) #vector del tiempo
m = Am*np.sinc(2*np.pi*Fm*(t-0.5*tf)/np.pi)#Am*np.cos(2*np.pi*Fm*t) #mensaje

Im = 0.8 # Im = peak(m(t))/Ac ->  Ac = peak(m(t))/Im

Ac = max(abs(m))/Im #amplitud de la portadora según índice de modulación
c = Ac*np.cos(2*np.pi*Fc*t) #portadora
print(Ac)
```

```python
# Graficamos la señales mensaje y portadora respectivamente

plt.plot(t,c,label='portadora')
plt.plot(t,m,label='mensaje')
plt.legend()
plt.show()
```

```python
# Cálculo de la señal modulada en AM y gráfica de la misma

y = (1+m/Ac)*c

plt.plot(t,y)
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
# Pasamos a calcular el espectro de frecuencia de la señales: mensaje, portadora y modulada
Xfc = np.fft.fft(c)
Xfm = np.fft.fft(m)
Xfy = np.fft.fft(y)

vfre = np.fft.fftfreq(len(c),1/Fs)

plt.plot(vfre,abs(Xfy),label='modulado')
plt.plot(vfre,abs(Xfm),label='mensaje')
plt.plot(vfre,abs(Xfc),label='portadora')
plt.legend()
plt.xlabel('')
plt.ylabel('')
plt.ylim([min(abs(Xfm)), 1.1*max(abs(Xfm))])
plt.show()
```

```python
import soundfile as sf # para instalar pip install soundfile
#lee archivos wav
for name in name_:
    nombre_out = "output.wav"
    x, fs = sf.read(name+'.wav')

    print('Frecuencia de muestreo %.2f[Hz]\naudio %s' % (fs,name+'.wav'))
```

```python
from IPython.display import Audio

xpro = x.copy() #copiar archivos para procesar
ti = 20  # tiempo incio a procesar seg
tf = 25  # tiempo final a procesar seg
xs = x[int(ti*fs):int(tf*fs),0] # Para un solo canal de audio
Audio(x[int(ti*fs):int(tf*fs)].T,rate=fs)
```

```python
#Señal mensaje en tiempo y frecuencia
t_v = np.arange(ti,tf,1/fs) # vector de tiempo
plt.plot(t_v,xs) #graficar señal mensaje en el tiempo
plt.xlabel('tiempo[s]')
plt.ylabel('m(t)')
plt.legend()
plt.show()
```

```python
# Cálculo de Fourier señal mensaje
Me_w = np.fft.rfft(xs) # axis=0 permite aplicar fft por cada columna de xpro
vf = np.fft.rfftfreq(xs.shape[0],1/fs) #se crea el vector de frecuencias
plt.plot(vf,abs(Me_w))#se grafica la magnitud
plt.title(r'Espectro señal mensaje')
plt.xlabel(r'f[Hz]',fontsize = 14)
plt.ylabel(r'|M[F]|',fontsize = 14)
plt.show()

plt.plot(vf, np.angle(Me_w))#fase del espectro
plt.xlabel(r'f[Hz]',fontsize = 14)
plt.ylabel(r'|M[F]|')
```

```python
# Se asume una frecuencia mayor a 12.000 Hz
Fc = 15000 # frecuencia portadora
Im = 1 # Im = peak(m(t))/Ac ->  Ac = peak(m(t))/Im
Ac = max(abs(xs))/Im #amplitud de la portadora según índice de modulación
c = Ac*np.cos(2*np.pi*Fc*t_v) #portadora

```

```python
#repoducir portadora
Audio(c,rate=fs)
```

```python
#Señal portadora en tiempo y frecuencia
plt.plot(t_v,c) #graficar señal mensaje en el tiempo
plt.xlabel('tiempo[s]')
plt.ylabel('c(t)')
plt.show()
```

```python
#Fourier señal portadora
C_w = np.fft.rfft(c) # axis=0 permite aplicar fft por cada columna de xpro
plt.plot(vf,abs(C_w))# se grafica la magnitud
plt.title(r'Espectro Portadora')
plt.xlabel(r'f[Hz]',fontsize = 14)
plt.ylabel(r'|M[F]|',fontsize = 14)
plt.show()


plt.plot(vf, np.angle(C_w))#fase del espectro
plt.xlabel(r'f[Hz]',fontsize = 14)
plt.ylabel(r'|M[F]|',fontsize = 14)
```

```python
# Señal modulada
y = (1+xs/Ac)*c #señal modulada en AM
```

```python
#repoducir modulada
Audio(y,rate=fs)
```

```python
plt.plot(t_v,y) #graficar señal mensaje en el tiempo
plt.xlabel('tiempo[s]')
plt.ylabel('y(t)')
plt.show()
```

```python
#Fourier señal modulada
Y_w = np.fft.rfft(y) # axis=0 permite aplicar fft por cada columna de xpro
plt.plot(vf,abs(Y_w))#se grafica la magnitud
plt.title(r'Espectro Modulada')
plt.xlabel(r'f[Hz]',fontsize = 14)
plt.ylabel(r'|Y[F]|',fontsize = 14)
plt.show()


plt.plot(vf, np.angle(Y_w))#fase del espectro
plt.xlabel(r'f[Hz]',fontsize = 14)
plt.ylabel(r'|Y[F]|',fontsize = 14)
```
